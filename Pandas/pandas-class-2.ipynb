```python
class Test:
    def __init__(self):
        self.value = 42

    def __getattribute__(self, name):
        print(f"Getting {name}")
        return super().__getattribute__(name)

obj = Test()
print(obj.value)  
```

```python
class MyDescriptor:
    def __set__(self, instance, value):
        print(f"Setting value to: {value}")
        instance._value = value

class MyClass:
    attr = MyDescriptor()

obj = MyClass()
obj.attr = 10  # Triggers __set__, prints "Setting value to: 10"
```
